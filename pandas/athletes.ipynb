## Pandas DataFrame

```python
import pandas as pd
```

### Read a CSV file

```python
athlete_df = pd.read_csv('../datasets/Athletes.csv', index_col='Rank')
```

```python
athlete_df.info()
```

### Pandas DataFrame Drop unwanted Columns

```python
athlete_df1 = athlete_df.drop(['Wikipedia Page', 'dbpedia Page', 'Image'], axis=1)
```

### Pandas DataFrame Sorting

```python
athlete_df2 = athlete_df1.sort_values(by=['Rank'], ascending=True)
```

#### Average height of Athletes based on Countries

```python
athlete_year_height_df = athlete_df2.groupby(['Nation'])['Height (cm)'].mean()
athlete_year_height_df
```

##### Pandas Plot

```python
athlete_year_height_df.plot(xlabel='Nation', ylabel='Height')
```
