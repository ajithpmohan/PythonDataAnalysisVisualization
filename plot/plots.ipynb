## Multiple Histogram

```python
flowers_df = sns.load_dataset('iris')
```

```python
flowers_df.sepal_width.unique()
```

```python
flowers_df.species.unique()
```

```python
setosa_df = flowers_df[flowers_df.species  == 'setosa']
```

```python
setosa_df.head()
```

```python
versicolor_df = flowers_df[flowers_df.species  == 'versicolor']
```

```python
versicolor_df.head()
```

```python
virginica_df = flowers_df[flowers_df.species  == 'virginica']
```

```python
virginica_df.head()
```

```python
plt.title('Distribution of sepal width')
plt.hist(setosa_df.sepal_width, bins=np.arange(2,5,0.25))
plt.hist(versicolor_df.sepal_width, bins= np.arange(2, 5, 0.25))
plt.hist(virginica_df.sepal_width, bins= np.arange(2, 5, 0.25))
plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

```python
plt.title('Distribution of sepal width')
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], bins=np.arange(2,5,0.25))
plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

```python
plt.title('Distribution of sepal width')
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], bins=np.arange(2,5,0.25), stacked=True)
plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

## HeatMap

```python
flights_df = sns.load_dataset('flights')
flights_df
```

```python
pivot_flights = flights_df.pivot(index='month', columns='year', values='passengers')
pivot_flights
```

```python
plt.title('Passenger Counts by Month and Year')
sns.heatmap(pivot_flights, fmt='d', annot=True, cmap='Blues');
plt.show()
```
